```python
img = plt.imread('image.jpeg')
img.shape
```

```python
img_resized = cv2.resize(img, (180,180), interpolation = cv2.INTER_AREA)
plt.imshow(img_resized)
```

```python
vector = img_resized.reshape(-1,3)
vector.shape
```

#### Note
- The only tunable parameter in Mean Shift algorithm is bandwith. It decides the clusters based upon local hills

```python
clustering = MeanShift(bandwidth=100,max_iter=1).fit(vector)
```

```python
labels = clustering.labels_
labels.shape
```

```python
centers = clustering.cluster_centers_
centers = centers.astype('uint8')
centers.shape
```

It automatically found 6 clusters

```python
output = centers[labels]
```

```python
result = output.reshape(img_resized.shape)
```

```python
plt.imshow(result)
```

### Observation
- As Bandwidth (window) size increased, number of clusters decreased. 
    - Why? - Possibly all hills within a big window are considered to be one
